```python
#importing pandas library
import pandas as pd
#read data from a file with path ('filename.csv')
df_full= pd.read_csv('Mex to US_Jun to Jul_ all data.csv',  encoding='ISO-8859-1')
#to view the first 10 rows of the dataset
df_full.head(10)
```

```python
df_full.info()
```

```python
df_full['Ddate'] = pd.to_datetime(df_full['Ddate'])
df_full['Collect Date'] = pd.to_datetime(df_full['Collect Date'])
df_full['Collect Date'] = df_full['Collect Date'].dt.date
```

```python
# Specify the particular date you want to filter on
target_min_date = pd.to_datetime('2023-06-01')
target_max_date = pd.to_datetime('2023-06-30')


# Filter the DataFrame based on the specified date
filtered_df = df_full[(df_full['Ddate'] >= target_min_date) & (df_full['Ddate'] <= target_max_date)]

# filter for specific routes:
filtered_df = filtered_df[(filtered_df['Airport Code - From'] == 'SJD') & (filtered_df['Airport Code - To'] == 'LAX')]

#filter for Alaska and Delta
filtered_df = filtered_df[(filtered_df['CXR'] == 'Delta Airlines') | (filtered_df['CXR'] == 'Alaska Airlines')]
filtered_df.shape
```

```python
df=filtered_df.iloc[:, :14]
df.head()
```

```python
df = df.groupby(['CXR','Days to Departure'])['Fare'].mean().reset_index()
```

```python
df_pivoted = df.pivot(index=['Days to Departure'], columns='CXR', values='Fare').reset_index()
df_pivoted.dropna(inplace=True)
```

```python
df_pivoted.head(10)
```

#Linear Regression Model - Predict Price Differentiation b/w Delta and Alaska Airlines

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Assuming df_pivoted is your DataFrame with columns 'Collect Date', 'Days to Departure', 'Alaska Airlines Fare', 'Delta Airlines Fare'

# Convert 'Days to Departure' to positive
#df_pivoted['Days to Departure'] = df_pivoted['Days to Departure'].abs()

# Feature engineering: Add 'Days to Departure' as a feature
X = df_pivoted[['Days to Departure', 'Delta Airlines']]
y = df_pivoted['Delta Airlines'] - df_pivoted['Alaska Airlines']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)

# Train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on test data
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Plot predictions
plt.figure(figsize=(10, 6))
plt.plot(X_test['Days to Departure'], y_test, label='Actual Difference', marker='o')
plt.plot(X_test['Days to Departure'], y_pred, label='Predicted Difference', linestyle='dashed', marker='o')
plt.xlabel('Days to Departure')
plt.ylabel('Delta - Alaska Price Difference ($)')
plt.title('Actual vs Predicted Fare Difference (Linear Regression Model)')
plt.legend()
plt.show()
```

#Simple Random Forest Regressor Model - Predict Price Differentiation b/w Delta and Alaska Airlines

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Assuming df_pivoted is your DataFrame with columns 'Collect Date', 'Days to Departure', 'Alaska Airlines Fare', 'Delta Airlines Fare'

# Convert 'Days to Departure' to positive
#df_pivoted['Days to Departure'] = df_pivoted['Days to Departure'].abs()

# Feature engineering: Add 'Days to Departure' as a feature
X = df_pivoted[['Days to Departure', 'Delta Airlines']]
y = df_pivoted['Delta Airlines'] - df_pivoted['Alaska Airlines']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)

# Train a Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict on test data
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Plot predictions
plt.figure(figsize=(10, 6))
plt.plot(X_test['Days to Departure'], y_test, label='Actual Difference', marker='o')
plt.plot(X_test['Days to Departure'], y_pred, label='Predicted Difference', linestyle='dashed', marker='o')
plt.xlabel('Days to Departure')
plt.ylabel('Delta - Alaska Price Difference ($)')
plt.title('Actual vs Predicted Fare Difference (Random Forest Regressor)')
plt.legend()
plt.show()
```
