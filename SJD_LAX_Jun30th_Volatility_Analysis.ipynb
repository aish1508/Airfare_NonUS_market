```python
#importing pandas library
import pandas as pd
#read data from a file with path ('filename.csv')
df_full= pd.read_csv('Mex to US_Jun to Jul_ all data.csv',  encoding='ISO-8859-1')
#to view the first 10 rows of the dataset
df_full.head(10)
```

```python
df_full.info()
```

```python
df_full['Ddate'] = pd.to_datetime(df_full['Ddate'])
df_full['Collect Date'] = pd.to_datetime(df_full['Collect Date'])
df_full['Collect Date'] = df_full['Collect Date'].dt.date
```

```python
# Specify the particular date you want to filter on
target_date = pd.to_datetime('2023-06-30')
#target_date = pd.to_datetime('2023-06-29')
#target_date = pd.to_datetime('2023-07-21')

# Filter the DataFrame based on the specified date
filtered_df = df_full[df_full['Ddate'] == target_date]

# filter for specific routes:
filtered_df = filtered_df[(filtered_df['Airport Code - From'] == 'SJD') & (filtered_df['Airport Code - To'] == 'LAX')]
filtered_df.shape
```

```python
df=filtered_df.iloc[:, :14]
df.head()
```

```python
df=df[['CXR','Collect Date','Fare','Days to Departure']]
df.head()
```

```python
df = df.groupby(['CXR','Collect Date','Days to Departure'])['Fare'].mean().reset_index()
```

```python
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')
df = df.sort_values(by=['CXR', 'Collect Date','Days to Departure'])
df['Daily_Return'] = df.groupby('CXR')['Fare'].pct_change()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with columns 'CXR', 'Days to Departure', and 'Fare'

# Convert 'Fare' to numeric
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')

# Calculate daily returns for each carrier
df['Daily_Return'] = df.groupby('CXR')['Fare'].pct_change()
```

```python
window_size = 7
```

```python
# Volatility
df['Volatility'] = df.groupby('CXR')['Daily_Return'].rolling(window=window_size).std().reset_index(level=0, drop=True)

# Set threshold for high CV
threshold = 0.20

# Identify periods of high CV
df['High_Volatility'] = (df['Volatility'] > threshold).astype(int)

# Specify colors for carriers
carrier_colors = {'Alaska Airlines': 'C0', 'Delta Airlines': 'C1', 'United Airlines': 'C3'}

# Plot Multiple Color Lines for Carriers
plt.figure(figsize=(10, 6))

for carrier in df['CXR'].unique():
    plt.plot(
        df[df['CXR'] == carrier]['Days to Departure'],
        df[df['CXR'] == carrier]['Volatility'],
        label=f' {carrier}',
        color=carrier_colors.get(carrier, 'grey')  # Default to grey if carrier color not specified
    )

# Highlight periods of high CV
plt.scatter(df[df['High_Volatility'] == 1]['Days to Departure'], df[df['High_Volatility'] == 1]['Volatility'], color='darkblue', label='High Volatility')

plt.xlabel('Days to Departure')
plt.ylabel('Volatility')
plt.title('Airline Fare Volatility Over Time')
plt.legend()
plt.show()
```

```python
# CV
df['CV'] = df.groupby('CXR')['Daily_Return'].rolling(window=window_size).std().reset_index(level=0, drop=True) / df.groupby('CXR')['Daily_Return'].rolling(window=window_size).mean().reset_index(level=0, drop=True)

# Set threshold for high CV
threshold = 100

# Identify periods of high CV
df['High_CV'] = (df['CV'] > threshold).astype(int)

# Specify colors for carriers
carrier_colors = {'Alaska Airlines': 'C0', 'Delta Airlines': 'C1', 'United Airlines': 'C3'}

# Plot Multiple Color Lines for Carriers
plt.figure(figsize=(10, 6))

for carrier in df['CXR'].unique():
    plt.plot(
        df[df['CXR'] == carrier]['Days to Departure'],
        df[df['CXR'] == carrier]['CV'],
        label=f' {carrier}',
        color=carrier_colors.get(carrier, 'grey')  # Default to grey if carrier color not specified
    )

# Highlight periods of high CV
plt.scatter(df[df['High_CV'] == 1]['Days to Departure'], df[df['High_CV'] == 1]['CV'], color='darkblue', label='High CV')

plt.xlabel('Days to Departure')
plt.ylabel('Coefficient of Variation (CV)')
plt.title('Airline Fare CV(coeff. of variation) Over Time')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with columns 'Days to Departure', 'Fare', 'Volatility', and 'CXR' (Carrier)

# Create a dictionary to store carrier-specific data and colors
carrier_data = {}
# Specify colors for carriers
carrier_colors = {'Alaska Airlines': 'C0', 'Delta Airlines': 'C1', 'United Airlines': 'C3'}

# Loop through unique carriers to calculate and store carrier-specific data
for carrier in df['CXR'].unique():
    carrier_df = df[df['CXR'] == carrier]
    carrier_data[carrier] = {
        'Days to Departure': carrier_df['Days to Departure'],
        'Fare': carrier_df['Fare'],
        'Volatility': carrier_df['Volatility']
    }

# Plotting for multiple carriers with specified colors
plt.figure(figsize=(12, 6))

for carrier, data in carrier_data.items():
    color = carrier_colors.get(carrier, 'black')  # Default to black if carrier color is not specified
    plt.plot(data['Days to Departure'], data['Fare'], label=f'{carrier}', alpha=0.7, color=color)
    plt.fill_between(data['Days to Departure'], data['Fare'] - data['Volatility']*data['Fare'], data['Fare'] + data['Volatility']*data['Fare'], alpha=0.15, color=color)

plt.xlabel('Days to Departure')
plt.ylabel('Fare ($)')
plt.title('Airline Fare Volatility (bands) Over Time for Multiple Carriers')
plt.legend()
plt.show()
```
